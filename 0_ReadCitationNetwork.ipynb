```python
venue_types = set()
venuedict_keys = Counter()
venuecount = 0
counter = 0

with fileinput.input("/home/singh_shruti/data/mag_aminer/citation_network/dblpv13.json") as f:
    for line in f:
        if counter == 0:
            counter += 1
            continue
        counter += 1
        try:
            line = line.strip()
            if line.startswith(",{"):
                line = line[1:]
            try:
                pdict = json.loads(line)
            except:
                pdict = ast.literal_eval(line)
                

            if "venue" in pdict:
                venuecount += 1
                venue_types.add(type(pdict["venue"]))
                venuedict_keys.update(list(pdict["venue"].keys()))
            else:
                print(counter)

        except Exception as ex:
            print(counter, ex)
```

```python
counter = 0

relevant_venue_fields = ['_id', 'raw', 'type', 'name_d', 'sid', 't', 'name_s', 'name', 'src']
venue_stats = {x: Counter() for x in relevant_venue_fields}

with fileinput.input("/home/singh_shruti/data/mag_aminer/citation_network/dblpv13.json") as f:
    for line in f:
        if counter == 0:
            counter += 1
            continue
        counter += 1
        
        if counter%500000 == 0:
            print(counter)
        
        try:
            line = line.strip()
            if line.startswith(",{"):
                line = line[1:]
            try:
                pdict = json.loads(line)
            except:
                pdict = ast.literal_eval(line)
                

            if "venue" in pdict:
                for k in pdict["venue"]:
                    if k in relevant_venue_fields:
                        venue_stats[k].update([pdict["venue"][k]])
            
        except Exception as ex:
            print(counter, ex)
```

```python
c_sens = ["NIPS", "NeurIPS", "ICLR", "ACL", "ICML", "NAACL", "EMNLP", "LREC", "EACL", "CVPR", "IJCAI", "AAAI", "IJCNLP", "TACL", "CoNLL", "ANLP", "COLING", "SIGKDD", "ICCV", "ECCV", "ICNN", "UAI", "IJCNN", "ICPR", "ICDAR", "CONLL"]

c_inses = ["conference on neural networks", "conference on pattern recognition", "conference on document analysis and recognition", "uncertainty in ai", "european conference on computer vision", "knowledge discovery", "computational linguistics", "computer vision and pattern recognition", "computational natural language learning", "learning representations", "neural information processing", "international conference on machine learning", "international conference on artificial intelligence", "international conference on computer vision", "semeval"]

counter = 0

network_dict = {}
edge_dict = {}

# relevant_venue_fields = ['_id', 'raw', 'type', 'name_d', 'sid', 't', 'name_s', 'name', 'src']
# venue_stats = {x: Counter() for x in relevant_venue_fields}

with fileinput.input("/home/singh_shruti/data/mag_aminer/citation_network/dblpv13.json") as f:
    for line in f:
        if counter == 0:
            counter += 1
            continue
        counter += 1
        
#         if counter == 6:
#             break
        if counter%500000 == 0:
            print(counter)
        
        try:
            line = line.strip()
            if line.startswith(",{"):
                line = line[1:]
            try:
                pdict = json.loads(line)
            except:
                pdict = ast.literal_eval(line)
                

            if "venue" in pdict:
                if "raw" in pdict["venue"]:
                    raw_venue_name = pdict["venue"]["raw"]
                    found = False
                    
                    for conf in c_sens:
                        if raw_venue_name.find(conf) > -1:
                            if "_id" in pdict and pdict["_id"]: 
                                if type(pdict["_id"])==str:
                                    network_dict[pdict["_id"]] = pdict
                                else:
                                    print(counter, pdict["_id"])
                            found = True
                            break
                    if not found:
                        for conf in c_inses:
                            if raw_venue_name.lower().find(conf) > -1:
                                if "_id" in pdict and pdict["_id"]: 
                                    if type(pdict["_id"])==str:
                                        network_dict[pdict["_id"]] = pdict
                                    else:
                                        print(counter, pdict["_id"])
                            found = True
                            break
                            
#                     if k in relevant_venue_fields:
#                         venue_stats[k].update([pdict["venue"][k]])
            
        except Exception as ex:
            print(counter, ex)
```
