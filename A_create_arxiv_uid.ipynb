## Create UID

```python
paper_id_titles = defaultdict(list)

with fileinput.input("data/arxiv_kaggle/arxiv-metadata-oai-snapshot.json") as f:
    for line in f:
        paper_dict = json.loads(line)
        
        title = paper_dict["title"]
        title_lower = title.lower()
        unaccented_title = unidecode.unidecode(title_lower)
        clean_title = re.sub('[\W_]', '', unaccented_title)
        
        aid = paper_dict['id']
        
        if clean_title:
            paper_id_titles[clean_title].append(aid)
        else:
            print(paper_dict['id'])
```

Find titles which are same for more than one paper

```python
c = 0

for k, v in paper_id_titles.items():
    if len(v) > 1:
        print(k, v)
        c += 1
print(c)
```

Find titles of length less than 20 characters. 
These will cause issues in mapping bibtex to titles as some are very common titles such as ispnp (is p=np), deepneuralnetworks, lstm

```python
c = 0
for k, v in paper_id_titles.items():
    if len(k) < 20:
        print(k, v)
        c += 1
print(c)
```

Assign a global sequence id to each title

```python
paper_global_id_dict = {}

seqid = 1
for k in paper_id_titles:
    if len(k) >= 20:
        paper_global_id_dict[k] = seqid
        seqid += 1
```

```python
with open("data/arxiv_paper_title_id_map.pkl", "wb") as f:
    pickle.dump(paper_global_id_dict, f)

with open("data/arxiv_paper_title_arxid_map.pkl", "wb") as f:
    pickle.dump(paper_id_titles, f)
```
